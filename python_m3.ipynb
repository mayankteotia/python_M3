```python
1. Explain the key features of Python that make it a popular choice for programming

Simple and Readable Syntax: Easy to learn and use.

Interpreted Language: No need for compilation.

Dynamically Typed: No need to declare variable types explicitly.

Extensive Libraries: Supports data science, AI, web development, etc.

Portability: Code runs across different platforms.

Open Source: Free to use, with strong community support.

Object-Oriented: Supports classes and objects.

Large Ecosystem: Libraries like NumPy, Pandas, TensorFlow, Flask, Django.

2. Describe the role of predefined keywords in Python and provide examples of how they are used in a program

Predefined keywords are reserved words in Python with special meanings.

They cannot be used as variable names or identifiers.

Examples:

if, else, for, while (control flow)

def (define a function)

class (define a class)

import (import a module)

Example:


if 5 > 3:
    print("5 is greater than 3")
3. Compare and contrast mutable and immutable objects in Python with examples

Mutable Objects: Can be changed after creation.

Examples: list, dict, set


my_list = [1, 2, 3]
my_list[0] = 10  # List is mutable
Immutable Objects: Cannot be changed after creation.

Examples: int, float, tuple, str


my_tuple = (1, 2, 3)
# my_tuple[0] = 10  # Error! Tuple is immutable
4. Explain the concept of type casting in Python with examples

Type Casting: Converting one data type to another.

Implicit (automatic) or explicit (manual) conversion.

Example:


x = 5  # integer
y = 2.5  # float
z = x + y  # Implicit casting to float

a = "123"
b = int(a)  # Explicit casting from string to int
5. Discuss the different types of operators in Python and provide examples of how they are used

Arithmetic Operators: +, -, *, /, //, %, **


print(5 + 3, 5 ** 2)
Comparison Operators: ==, !=, >, <, >=, <=

Logical Operators: and, or, not

Assignment Operators: =, +=, -=

Bitwise Operators: &, |, ^, ~, <<, >>

Membership Operators: in, not in

Identity Operators: is, is not

6. How do conditional statements work in Python? Illustrate with examples
Conditional statements allow decision-making:


age = 18
if age >= 18:
    print("Adult")
else:
    print("Minor")
You can use:

if

if-else

if-elif-else for multiple conditions.

7. Describe the different types of loops in Python and their use cases with examples

for loop: Iterates over a sequence.


for i in range(5):
    print(i)
while loop: Repeats while a condition is True.


count = 0
while count < 5:
    print(count)
    count += 1
break and continue can control loop execution.

Use cases:

For loops: When you know how many times to iterate (like over a list).

While loops: When you want to repeat until a condition changes (like waiting for user input).

```
